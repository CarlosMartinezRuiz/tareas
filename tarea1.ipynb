Agrupamos:

```python
semanal = df.groupby('SEMANA').size().reset_index(name='MUERTES')
```

```python
semanal
```

```python
semanalgripe = dfgripe.groupby(['GLOSA_CATEGORIA_DIAG1','SEMANA']).size().reset_index(name='MUERTES')
```

```python
semanalcovid = dfcovid.groupby(['SEMANA']).size().reset_index(name='MUERTES')
```

```python
semanalcovid
```

Ahora con los 3 grupos graficamos para concluir

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(semanalcovid['SEMANA'],semanalcovid['MUERTES'],label='COVID19')
plt.plot(semanal['SEMANA'],semanal['MUERTES'], label='TOTAL')
plt.legend(loc="upper left") 
plt.xlabel('N° de Semana')
plt.ylabel('Fallecidos')
plt.title('Muertes totales en chile comparado con COVID')
plt.show()
```

```python
plt.plot(semanalcovid['SEMANA'],semanalcovid['MUERTES'],label='COVID19')
plt.plot(semanalgripe['SEMANA'],semanalgripe['MUERTES'], label='GRIPE')
plt.legend(loc="upper left") 
plt.xlabel('N° de Semana')
plt.ylabel('Fallecidos')
plt.title('Muertes *de gripe* en Chile comparada con COVID')
plt.show()
```

Podemos observar teniendo en cuenta ambos graficos que el Covid en cuanto a las muertes dentro de los ultimos años ha sido mucho mas relevante que la gripe por lo que podemos decir que es mas mortifero al dia de hoy pero hay que tener en cuenta que las gripes comunes ya llevan años y la mayor parte de la poblacion tiene vacunas ante estas cosa que aun no sucede con el COVID por lo que habria que comparar los datos de COVID de este ultimo año vs los datos de la llegada de "viejas" gripes en su momento para determinar si finalmente correspode a una gripe comun en comparativa o mas bien es una gripe mas "mortifera".
